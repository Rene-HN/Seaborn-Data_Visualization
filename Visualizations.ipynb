###5. Is there a correlation between lotarea and the saleprice?
Although the price of unit of area would vary depending on the proximity to important places, there is still some correlation between the size of a lot and the sale price of a house.

```
houseprice_corr = houseprices_df[['saleprice', 'lotarea']]

sns.pairplot(houseprice_corr);
```

```
plt.figure(figsize = (11,5))
sns.scatterplot(x="lotarea", y="saleprice", data = houseprice_corr)
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.title('Sale Price versus Lot Area')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium'  
)
plt.show()
```

```
plt.figure(figsize = (11,5))
sns.scatterplot(x="saleprice", y="lotarea", data = houseprice_corr)
plt.xlabel('Sale Price')
plt.ylabel('Lot  Area')
plt.title('Sale Price versus Lot Area')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium'  
)
plt.show()
```
